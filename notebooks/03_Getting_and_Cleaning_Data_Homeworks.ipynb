1. Register an application with the Github API [here](https://github.com/settings/applications). 
2. Access the API to get information on [your instructor's repositories](https://api.github.com/users/jtleek/repos). 
3. Use this data to find the time that the `datasharing` repo was created.  
   What time was it created? 

[This tutorial](https://github.com/hadley/httr/blob/master/demo/oauth2-github.r) may be useful. You may also need to run the code in the base R package and not R studio.
* 2014-02-06T16:13:11Z
* 2012-06-21T17:28:38Z
* 2013-11-07T13:25:07Z
* 2014-03-05T16:11:46Z

````python
names(s <- Sys.getenv())
````

````python
# httpuv package required (by httr) to capture OAuth credentials.
install.packages('httpuv')
````

````python
# https://github.com/hadley/httr/blob/master/demo/oauth2-github.r

library(httr)
library(httpuv)
library(jsonlite)
````

````python
# 1. Find OAuth settings for github:
#    http://developer.github.com/v3/oauth/

githubOAuth = oauth_endpoints("github")
````

````python
# 2. Register an application at https://github.com/settings/applications;
#    Use any URL you would like for the homepage URL (http://github.com is fine)
#    and http://localhost:1410 as the callback url
#
#    Insert your client ID and secret below - if secret is omitted, it will
#    look it up in the GITHUB_CONSUMER_SECRET environmental variable.

myApp <- oauth_app('R-getting-data ', 
                   Sys.getenv('GITHUB_CLIENT_ID'), 
                   Sys.getenv('GITHUB_CLIENT_SECRET'))
````

````python
# 3. Get OAuth credentials
github_token <- oauth2.0_token(githubOAuth, myApp, cache = TRUE)
````
